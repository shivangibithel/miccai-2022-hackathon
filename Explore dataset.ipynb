```python
# create single row for 1 patient by averaging across the column values in each table

final_df = pd.DataFrame(columns = list_headers)
final_df

# taking mean as a representation of patients complete data over the time and adding to final_df
for i in range(250):
    a = pd.read_csv(list_bloodvalues_files[i], index_col=0)
    b = pd.read_csv(list_radiology_files[i], index_col=0)
    c = pd.read_csv(list_temp_files[i], index_col=0)
    d = pd.read_csv(list_outcome_files[i], index_col=0)

    col_lis1 = list(a)
    col_lis2 = list(b)
    col_lis3 = list(c)
    col_lis4 = list(d)
    
    a_mean = a.mean()
    b_mean = b.mean()
    c_mean = c.mean()
    d_mean = d.mean()
    
    dic = {}
    
    for j in col_lis1:
        dic[j] = a_mean[j]
    for j in col_lis2:
        dic[j] = b_mean[j]
    for j in col_lis3:
        dic[j] = c_mean[j]
    for j in col_lis4:
        dic[j] = d_mean[j]
#     print(dic)
    final_df = final_df.append(dic, ignore_index=True)
#     break
final_df
```
